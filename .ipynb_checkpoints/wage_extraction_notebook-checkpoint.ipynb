```python
def clean_and_split_str(txt):
    #strip_special_chars = re.compile("[^A-Za-z0-9#]+")
    translator = str.maketrans('', '', string.punctuation)
    txt = txt.translate(translator)
    txt = re.sub('\s+', ' ', txt).strip()
    txt = txt.lower()
    txt = txt.split(' ')
    return txt

def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)
```

```python
df['clean'] = [clean_and_split_str(i) for i in df.ocr]
```

** Line Extraction **

The loop below searches for occupations in the cleaned articles. Here a list is used that includes the HISCO occupation and OCR-'alternatives' aquired through cosine similarity scores (in a vector space model) and string distance measuring. Once an occupation is identified, a region 'around' it (L12:R18) is selected. If the word 'loon' or the 'f' ('gulden') sign occurs within this region, the region is added to a list, along with the occupation.

```python
## We take the clean list of articles from the dataframe and detect occupations using the list
## We select only the occupation-windows that contain the word 'loon' or the 'ƒ' sign
## We pick a window of 2 left of the occupation, and 12 right of the occupation

lo = list()
ctr = 0

for c,i in enumerate(df.clean):
    ctr += 1
    if any(e in i for e in list_words):
        
        occ = list()
        
        for w in i:
            if w in list_words:
                occ.append(w)
        
        for o in occ:
            ind = i.index(o)
            sl = i[ind-12:ind+18]
            sl = ' '.join(sl)
            if 'loon ' in sl or 'ƒ' in sl:
                lo.append([o,sl])
        
    if ctr % 10000 == 0:
        print(str(round(ctr / len(df.clean) * 100)) + "%")
```

** Inspection of the Data **

Before extracting wages, we need to have an idea what kind of occupations inhabit the advertisements. For this to analyze, I first inspect whether we have enough data to have most occupations represented. I do this by plotting the number of lines against the productivity of the occupations.



```python
## Check when we have a large enough sample

ocprod = pd.DataFrame()

list_occ = []

for c,i in enumerate(lo):
    list_occ.append(i[0])
    iteration = c
    productivity = len(list(set(list_occ)))
    
    t = pd.DataFrame([iteration, productivity]).T
    ocprod = ocprod.append(t)

ocprod.columns = ['iteration', 'productivity']
p = ocprod.plot.line(x='iteration', y='productivity')
```

```python
occdist = pd.DataFrame([list(dict(Counter(list_occ)).keys()), list(dict(Counter(list_occ)).values())]).T
occdist.columns = ['occ', 'count']
occdist = occdist[occdist['count'] > 50]
```

```python
locw = list()

for oc,st in lo:
    tmp = st.split()
    
    ## double check if loon is in section
    if "loon" not in tmp:
        continue
    
    #get index of loon
    il = tmp.index("loon")

    # if loon is last word of section, discard
    if il == len(tmp) -1:
        continue
    selection = tmp[il+1:il+3] 
    locw.append([oc," ".join(selection)])
```

```python
def CloseRangeWage(string):
    # Split string
    s = string.split()
    
    # Check if we have separate f-sign, no f-sign or one f-sign combined with wages
    
    if 'f' in s or 'ƒ' in s:
        type_s = "separate_f"
    else:
        type_s = "not_separate_f"
        
    if type_s != "separate_f":
        if any("f" in x for x in s) or any("ƒ" in x for x in s):
            type_s = "internal_f"
        else:
            type_s = "no_f"
    
    
    
    
    

    return type_s
```

```python
### Check in how many articles that mention an occupation, an instance of "wage" occurs

for occupation in list(set([o for o,v in locw])):

    c_occupation = 0
    c_loon = 0

    for oc,st in lo:

        st = st.split()

        if occupation in st:
            c_occupation += 1
        if "loon" in st and occupation in st:
            c_loon += 1

    print(occupation + " | " + str(c_loon / c_occupation * 100) + " | " + str(c_loon))
```
