```python
import os, glob, pandas as pd
from gensim.models import KeyedVectors
from gensim.models import Word2Vec
```

```python
os.chdir("C://Users//Ruben//Documents//DHLab")
df = pd.read_csv("classified_ads.csv")
```

```python
os.chdir("C://Users//Ruben//Documents//GitHub//TheForeign//SCC//output-data")
model_whole = KeyedVectors.load_word2vec_format('model-whole.bin', binary = True)
```

```python
stopz = [word.split() for word in open("C:\\Users\\Ruben\\Documents\\Artikelen\\Joris\\stopwords-nl.txt", 'r', encoding = "utf-8").readlines()]
stopz = [item for sublist in stopz for item in sublist]
```

```python
dfr = df[df.class_name == 1]

dict_wage_context = dict()

for c,i in enumerate(dfr.clean_art):
    
    if 'ƒ' in i:
        article = i.split(' ')
        
        list_wca = list()
        
        for count,word in enumerate(article):
            if 'ƒ' in word:
                tmp = ' '.join(article[count-12:count+12])
                list_wca.append(tmp)
        dict_wage_context.update({dfr.index[c]:list_wca})
```

```python
list_wage_context = list()
for key, value in dict_wage_context.items():
    list_wage_context.append(value)
```

```python
list_wage_context = [item for sublist in list_wage_context for item in sublist]
```

```python
def DistanceMatrix(list_sentences):
        
    total_list = list()
    
    for sentence in list_sentences:
        sentence = sentence.split()
        sentence = [w for w in sentence if w not in stopz]
        
        list_sent = list()
        
        for sent2 in list_sentences:
            sent2 = sent2.split()
            sent2 = [w for w in sent2 if w not in stopz]
            tmp = model_whole.wmdistance(sentence, sent2)
            list_sent.append(tmp)
        
        total_list.append(list_sent)
    df = pd.DataFrame(total_list, columns = list_sentences, index = list_sentences)
    return df
```

```python
test_df = list_wage_context[0:100]
test_df = list(set(test_df))
distmat = DistanceMatrix(test_df)
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np
from adjustText import adjust_text

%matplotlib inline

def tsne_plot_dist(distance_matrix):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    ldm = len(distmat) - 1
    
    for i in range(0,ldm) :
        tokens.append(np.array(distance_matrix.iloc[i,0:ldm]))
        labels.append(distance_matrix.index[i])
    
    tsne_model = TSNE(perplexity=10, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    return [labels, new_values]
    
list_results = tsne_plot_dist(distmat)
new_values = list_results[1]
labels = list_results[0]

x = []
y = []

for value in new_values:
    x.append(value[0])
    y.append(value[1])
        
plt.figure(figsize=(50, 40)) 
scatter = [plt.scatter(x[i], y[i]) for i in range(len(x))]
texts = [plt.annotate(labels[i], xy=(x[i], y[i]), xytext=(5, 2),textcoords='offset points', ha='right',va='bottom', fontsize=20) for i in range(len(x))]
adjust_text(texts)
plt.show()
```
