```python
from nltk.tokenize import sent_tokenize
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import glob
import os
from os import path
import string
import numpy
import re
from __future__ import absolute_import, division, print_function
import pandas as pd

import gensim
```

```python
def clean_and_split_str(txt):
    #strip_special_chars = re.compile("[^A-Za-z0-9#]+")
    translator = str.maketrans('', '', string.punctuation)
    txt = txt.translate(translator)
    txt = re.sub('\s+', ' ', txt).strip()
    txt = txt.lower()
    txt = txt.split(' ')
    return txt
```

```python
sentences = list()

for np in ['roni', 'nvdd', 'algh']:
    os.chdir("C://Users//Ruben//Downloads//metadata//" + np + '_set')
    
    df = pd.read_csv('1880-1890.csv', sep = '\t')
    
    for art in list(df.ocr):
        t = clean_and_split_str(art)
        sentences.append(t)
```

```python
os.chdir("C://Users//Ruben//Documents//DHLab")

model = Word2Vec(sentences, size=150, window=10, iter = 8, min_count=2, workers=3)
model.train(sentences,total_examples=len(sentences),epochs=10)
model.wv.save_word2vec_format("modelw2v.bin", binary=True)
```

```python
os.chdir("C://Users//Ruben//Documents//GitHub//dhl-ads//resources")

with open('list_occupations.txt', encoding = 'utf-8') as f:
    list_words = f.read().splitlines()
    
list_words = list(set(list_words))
```

```python
import editdistance
```

```python
list_words = [w for w in list_words if len(w.split(' ')) == 1]

dict_alt = dict()

for i in list_words:
    
    if i in list(model.wv.vocab.keys()):
        ls = list(dict(model.most_similar(i, topn=20)).keys())
    else:
        continue
    
    la = list()
    for w in ls:
        if editdistance.eval(i,w) < 3:
            la.append(w)
    dict_alt.update({i:la})
```
