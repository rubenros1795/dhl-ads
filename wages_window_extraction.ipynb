```python
def MSW(word,n):
    t = dict(model_whole.most_similar(word, topn=n))
    return list(t.keys())

def GetList(term):
    mtf = list()

    for word in MSW(term, 5):
    
        word_list = list()
    
        lw = MSW(word, 5)
    
        for w in lw:
            tmp = MSW(w, 5)
            for i in tmp:
                word_list.append(i)
    
        for x in word_list:
            mtf.append(x)
    return sorted(list(set(mtf)))

def GetListMultipleKW(list_of_terms):
    
    master_list = list()
    
    for term in list_of_terms:
        tmp = GetList(term)
        for t in tmp:
            master_list.append(t)
    
    
    master_list = pd.DataFrame(master_list, columns = ['word'])
    vc = master_list.word.value_counts() == len(list_of_terms)
    vc = vc[vc]
    master_list.loc[master_list.word.isin(vc.index)]
    
    return(list(master_list.word))
```

```python
dict_wage_context = dict()

for c,i in enumerate(df.cleaned):
    
    if 'ƒ' in i or 'loon ' in i:
        article = i.split(' ')
        
        list_wca = list()
        
        for count,word in enumerate(article):
            if 'ƒ' in word:
                tmp = ' '.join(article[count-12:count+12])
                list_wca.append(tmp)
        dict_wage_context.update({df.index[c]:list_wca})
```

```python
list_wage_context = list()
for key, value in dict_wage_context.items():
    list_wage_context.append(value)
list_wage_context = [item for sublist in list_wage_context for item in sublist]
list_wage_context = list(set(list_wage_context))
list_wage_context = [l.split() for l in list_wage_context]
```

```python
os.chdir("C://Users//Ruben//Documents//GitHub//TheForeign//SCC//output-data")
model_whole = KeyedVectors.load_word2vec_format('model-whole.bin', binary = True)
```

```python
vl = np.average(model_whole[GetListMultipleKW(['huishoudster', 'timmerman', 'dienstbode', 'slager', 'beambte', 'arbeider', 'smid'])], axis=0)
model_whole.add('occ_av',vl,replace=False)
```

```python
for i in list_wage_context: 
    list_scores = list()
    for x in i:
        if x in list(model_whole.vocab.keys()):
            sim =  model_whole.similarity(x, 'occ_av')
            tmp = [x, sim]
            list_scores.append(tmp)
    tmp = pd.DataFrame(list_scores, columns = ['word', 'score'])
    #tmp = tmp.sort_values(['score'], ascending = False)
    tmp = tmp[tmp.score > 0.7]
    tmp = list(tmp.word)
    if len(tmp) > 0:
        print(' | '.join(tmp))
```
