```python
ol = []
os.chdir('C:\\Users\\Ruben\\Documents\\GitHub\\dhl-ads\\evaluation\\txt')

count = 0
for c,row in enumerate(df['oc']):
    list_oc = ast.literal_eval(row)
    
    for o in list_oc:
        oc = o.split("_")[0]
        
        ind = int(o.split("_")[1])
        
        window = df['clean'][c].split(' ')
        
        left = ind - 12
        if left < 0:
            left = 0
            
        right = ind + 40
        if right > len(window):
            right = len(window)
        
        window = window[left:right]
        window = " ".join(window)
        
        fn = oc + "_" + df['id'][c]
        fn = fn.replace('/',"")
        fn = fn.replace(':',"_")
        fn = fn + '.txt'
        
        job_ad_indicator = ['gevraagd','loon','salaris','geplaatst', 'men verlangt', 'jaarwedde']
        
        if any(word in window for word in job_ad_indicator): 
            count += 1
            ol.append(oc)
            print("{} | {}".format(oc,ind))
            print(window)
            print('=================')
            fn = str(count) + "_" + fn
            with open(fn, 'w', encoding = 'utf-8') as the_file:
                the_file.write(window)
```
