```python
import os, glob, pandas as pd
from gensim.models import KeyedVectors
from gensim.models import Word2Vec
from sklearn.feature_extraction.text import TfidfVectorizer
import string, re
```

```python
os.chdir("C://Users//Ruben//Documents//DHLab")
df = pd.read_csv('test_data.csv', sep = '\t')
```

```python
from sklearn.externals import joblib
loaded_model = joblib.load('lsvc_classifier.pkl')

def clean_and_split_str(txt):
    #strip_special_chars = re.compile("[^A-Za-z0-9#]+")
    translator = str.maketrans('', '', string.punctuation)
    txt = txt.translate(translator)
    txt = re.sub('\s+', ' ', txt).strip()
    txt = txt.lower()
    return txt
```

```python
df['cleaned'] = [clean_and_split_str(art) for art in df.ocr]
```

```python
tfidf_vec = TfidfVectorizer(min_df=1)
```

```python
tfidf_vec.transform(df.cleaned)
```

```python
dfr = df[df.class_name == 1]

dict_wage_context = dict()

for c,i in enumerate(dfr.clean_art):
    
    if 'ƒ' in i:
        article = i.split(' ')
        
        list_wca = list()
        
        for count,word in enumerate(article):
            if 'ƒ' in word:
                tmp = ' '.join(article[count-12:count+12])
                list_wca.append(tmp)
        dict_wage_context.update({dfr.index[c]:list_wca})
```

```python
list_wage_context = list()
for key, value in dict_wage_context.items():
    list_wage_context.append(value)
```

```python
list_wage_context = [item for sublist in list_wage_context for item in sublist]
```
